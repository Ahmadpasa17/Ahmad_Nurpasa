### Cleaning Data

Menghapus kolom data yang tidak diperlukan dari day_df dan hour_df

Merubah nama kolom agar mudah dibaca dan dimengerti

```python
hour_df.rename(columns={"dteday":'date', "mnth":'month', 'hr':'hour', 'weekday':'day', 'yr':'year', 'weathersit': 'weather_situation', 'cnt': 'count'}, inplace=True)
day_df.rename(columns={"dteday":'date', "mnth":'month', 'hr':'hour', 'weekday':'day' , 'yr':'year', 'weathersit': 'weather_situation', 'cnt': 'count'}, inplace=True)
```

## Visualization & Explanatory Analysis

### Pertanyaan 1: Pada tahun 2011, rental sepeda terbanyak dan terendah terjadi pada musim apa?

```python
season = ('fall', 'summer', 'winter', 'spring')
values = [419650, 347316, 326137, 150000]
colors = ('#8B4513', '#FFF8DC', '#93C572', '#E67F0D')
explode = (0.1, 0, 0, 0)

plt.pie(
    x=values,
    labels=season,
    autopct='%1.1f%%',
    colors=colors,
    explode=explode
)
plt.show()
```

### Pertanyaan 2: Pada tahun 2012, rental sepeda terbanyak dan terendah terjadi pada bulan apa?

```python
month = ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December']
values = [96744, 103137, 164875, 174224, 195865, 202830, 203607, 214503, 218573, 198841, 152664, 123713]

plt.figure(figsize=(10, 6))
plt.bar(month, values, color='skyblue')
plt.xlabel('Month')
plt.ylabel('Values')
plt.title('Monthly Distribution')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

## Conclusion

- Conclution pertanyaan 1 : Pada tahun 2011, rental sepeda terbanyak terjadi pada musim fall dengan presentase 33.8% atau sebanyak 419650. Dan, rental sepeda terendah terjadi pada musim spring dengan persentase 12.1% atau 150000.
- Conclution pertanyaan 2 : Pada tahun 2012, rental sepeda terbanyak terjadi di bulan September dan terendah pada bulan January
